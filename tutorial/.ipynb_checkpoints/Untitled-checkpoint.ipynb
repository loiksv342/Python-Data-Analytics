```python
x_values_female = [0, 1, 2]  # Punkty na osi X dla kobiet
x_values_male = [0, 1, 2]    # Punkty na osi X dla mÄ™Å¼czyzn

# WartoÅ›ci na osi Y dla kobiet i mÄ™Å¼czyzn
y_values_female = [120, 110, 100]  # RÃ³Å¼ne wartoÅ›ci, aby linia mogÅ‚a siÄ™ przeciÄ…Ä‡
y_values_male = [123, 115, 105]    # RÃ³Å¼ne wartoÅ›ci, aby linia mogÅ‚a siÄ™ przeciÄ…Ä‡

# Tworzymy wykres liniowy
plt.plot(x_values_female, y_values_female, label='Female', marker='o', linestyle='-')
plt.plot(x_values_male, y_values_male, label='Male', marker='o', linestyle='-')

# Ustawiamy etykiety osi X

# Dodajemy etykiety i tytuÅ‚
plt.xlabel('X Axis (Arbitrary Points)')
plt.ylabel('Math Score')
plt.title('Math Score Comparison by Gender')
plt.legend()
plt.grid(True)  # Opcjonalnie: Dodanie siatki dla lepszej wizualizacji
plt.show()
```
