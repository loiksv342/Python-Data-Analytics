### Question 1: Does Sleep Duration affect Stress Level

```python
sleep_agg = df.groupby('Sleep Duration')['Stress Level'].mean().reset_index()

plt.figure(figsize=(14, 7))

plt.plot(sleep_agg['Sleep Duration'], sleep_agg['Stress Level'], marker='o', color='green')

plt.title('Line Plot of Average Stress Level by Sleep Duration', fontsize=18)
plt.xlabel('Sleep Duration (hours)', fontsize=16)
plt.ylabel('Average Stress Level', fontsize=16)

plt.grid(True)
plt.show()
```

### Question 2: Which Occupations Have the Highest Frequency of Sleep Disorders?

```python
df_clean = df.dropna(subset=['Occupation', 'Sleep Disorder'])

plt.figure(figsize=(14, 7))
sns.countplot(x='Occupation', hue='Sleep Disorder', data=df_clean, palette='Set1')

plt.title('Distribution of Sleep Disorder Across Occupations', fontsize=18)
plt.xlabel('Occupation', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.xticks(rotation=45)  

plt.show()
```

### Question 3: How Are BMI Categories Distributed Across Different Genders

```python
df_clean = df.dropna(subset=['Gender', 'BMI Category'])

plt.figure(figsize=(14, 7))

sns.countplot(x='Gender', hue='BMI Category', data=df_clean, palette='Set1')

plt.title('Distribution of BMI Categories by Gender', fontsize=18)
plt.xlabel('Gender', fontsize=18)
plt.ylabel('Count', fontsize=18)

plt.show()
```

## 3. Correlation between Stress Level, Quality of Sleep, Age, Physical Activity Level

```python
matrix_columns = df[['Stress Level', 'Quality of Sleep', 'Age', 'Physical Activity Level']]  

correlation_matrix = matrix_columns.corr()  

plt.figure(figsize=(10, 8))  

sns.heatmap(correlation_matrix, annot=True, linewidths=0.5, cmap='coolwarm')  

plt.title('Correlation between Stress, Sleep Quality, Age, and Physical Activity', fontsize=18)  

plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)  

plt.show()  
```

### Question 4: How Does Physical Activity Level Vary by Gender and Age Group?

```python
bins = [25, 35, 45, 55]  
labels = ['26-35', '36-45', '46-55']  
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)  

df = df.dropna(subset=['Age Group'])  

g = sns.catplot(  
    x='Age Group',  
    y='Physical Activity Level',  
    hue='Gender',  
    data=df,  
    kind='bar',  
    height=5,  
    aspect=1.5  
)  

g.fig.suptitle('Physical Activity Level by Gender and Age Group', fontsize=16)  
g.set_axis_labels('Age Group', 'Physical Activity Level')  
g.legend.set_title('Gender')  

plt.show()  
```
