```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'covid.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

#### Question 1: Which WHO Region Has The Most Confirmed Cases?

```python
cases = df.groupby('WHO Region')['Confirmed'].sum()
plt.style.use('ggplot')

cases_sorted = cases.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
cases_sorted.plot(kind='bar', color='skyblue',)

plt.title('Total Confirmed Cases by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Total Confirmed Cases')
```

#### 1 Week Change in Confirmed Cases by WHO Region

```python
df_grouped = df.groupby('WHO Region')['1 week change'].sum().reset_index()

df_grouped = df_grouped.sort_values(by='1 week change', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(df_grouped['WHO Region'], df_grouped['1 week change'], color='skyblue')

plt.xlabel('WHO Region')
plt.ylabel('1 Week Change')
plt.title('1 Week Change in Confirmed Cases by WHO Region')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linewidth=0.7)
plt.tight_layout()  
plt.show()
```

#### Top 10 Countries of active cases

```python
top_10 = df.groupby('Country/Region')['1 week change'].sum().head(10)
top_10.sort_values(ascending = 1)

plt.figure(figsize=(10, 8))
top_10.plot.bar()
plt.title('Top 10 Countries by Active Cases')
plt.show()
```

```python
region = df[df['WHO Region'] == 'Americas'].copy()

region['Recovered per Case'] = region['Recovered'] / region['Confirmed'] * 100

region['Date'] = pd.to_datetime(region['Date'])

end_date = pd.to_datetime('2023-12-31')

start_date = end_date - pd.DateOffset(months=6)

filtered_region = region[(region['Date'] >= start_date) & (region['Date'] <= end_date)]

plt.figure(figsize=(12, 6))

plt.plot(filtered_region['Date'], filtered_region['Recovered per Case'], marker='o', linestyle='-', color='b', label='Recovery Ratio')

dates_ordinal = filtered_region['Date'].map(pd.Timestamp.toordinal)
z = np.polyfit(dates_ordinal, filtered_region['Recovered per Case'], 1)
p = np.poly1d(z)
plt.plot(filtered_region['Date'], p(dates_ordinal), linestyle='--', color='r', label='Trend Line')

plt.title('Recovery Ratio to Confirmed Cases in the Americas Region')
plt.xlabel('Date')
plt.ylabel('Recovery Ratio (%)')
plt.grid(True)
plt.legend()

plt.xticks(rotation=45) 

plt.tight_layout()
plt.show()
```

## Visualization of Recovery Ratio to Confirmed Cases in the Americas Region
The following steps are performed:

1. **Load the Data:**
   We load the COVID-19 dataset from a CSV file.

2. **Filter Data for the Americas Region:**
   We extract data specific to the "Americas" region.

3. **Calculate Recovery Ratio:**
   We compute the recovery ratio as the percentage of recovered cases out of confirmed cases.

4. **Date Range Filtering:**
   We select data from the last 6 months to focus on recent trends.

5. **Plotting:**
   - We create a line chart showing the recovery ratio over time.
   - We add a trend line to visualize the general trend.

6. **Formatting:**
   - We rotate the date labels on the x-axis to make them more readable.
   - We adjust the plot layout to ensure everything fits well.

The chart will help  understand how the recovery ratio has changed over time and identify any significant trends.

```python
df['Date'] = pd.to_datetime(df['Date'])

europe_data = df[df['WHO Region'] == 'Europe']

end_date = pd.to_datetime('2023-12-31')
start_date = end_date - pd.DateOffset(months=6)
filtered_data = europe_data[(europe_data['Date'] >= start_date) & (europe_data['Date'] <= end_date)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Date'], filtered_data['Active'], marker='o', linestyle='-', color='b', label='Active Cases')

plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=10))  # Ustaw maksymalną liczbę etykiet na osi X

plt.xlabel("Date")
plt.ylabel("Number of Active Cases")
plt.title("Number of Active COVID-19 Cases in Europe Over Time")
plt.grid(True)
plt.legend()

plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

## Visualizing Active COVID-19 Cases in Europe

In this section, we’ll create a plot to visualize the number of active COVID-19 cases in Europe. Here’s a straightforward guide on how we’ll do it:

1. **Load the Dataset**

4. **Narrow Down the Date Range**  
   To make our plot clearer, we’ll limit the data to the last 6 months. This way, we can focus on recent trends without being overwhelmed by too much data.

5. **Create the Plot**  
   - We’ll plot the number of active cases over time using a line plot.
   - To make the x-axis more readable, we’ll adjust the date labels.
   - We’ll add labels, a title, and a legend to ensure the plot is informative and easy to understand.

6. **Display the Plot**  
   Finally, we’ll rotate the x-axis labels if needed and ensure the plot looks clean and professional.

By the end, we’ll have a clear visual representation of how the number of active COVID-19 cases in Europe has changed over the past few months.
