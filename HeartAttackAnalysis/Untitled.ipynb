```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'heart_attack_dataset.csv')
df.head(5)
```

### Question 1: Who smokes the most by age group

```python
grouped = df.groupby(['Age', 'Smoking Status']).size().reset_index(name='Counts')
plt.figure(figsize=(12, 6))

sns.lineplot(data=grouped, x='Age', y='Counts', hue='Smoking Status', marker='x')

plt.title('Smoking Status by Age Group', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of People', fontsize=14)

plt.legend(title='Smoking Status', title_fontsize='13', fontsize='11')

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a bar plot showing the count of treatments by gender
sns.countplot(data=df, x='Gender', hue='Treatment')

plt.title('Distribution of Treatments by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Adjusting legend for better readability
plt.legend(title='Treatment', title_fontsize='13', fontsize='11')

# Show the plot
plt.show()
```

```python
data = df.groupby('Age')[['Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']].mean().reset_index().round(1)

# Define x values and plot
x = data['Age']

plt.figure(figsize=(18, 8))

# Plot Blood Pressure and Cholesterol
plt.plot(x, data['Blood Pressure (mmHg)'], label='Blood Pressure', marker='o')
plt.plot(x, data['Cholesterol (mg/dL)'], label='Cholesterol', marker='o')

# Add labels and title
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Value', fontsize=14)
plt.title('Average Blood Pressure and Cholesterol Levels by Age', fontsize=16)

# Adjust x-ticks to match the age values
plt.xticks(x, rotation=45)

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grupowanie według 'Age' i 'Gender', a następnie obliczanie średnich wartości
data = df.groupby(['Age', 'Gender'])[['Blood Pressure (mmHg)', 'Cholesterol (mg/dL)']].mean().reset_index().round(1)

# Wykres słupkowy dla ciśnienia krwi
plt.figure(figsize=(18, 8))
sns.barplot(data=data, x='Age', y='Blood Pressure (mmHg)', hue='Gender', palette='tab10', errorbar=None)

# Dodanie tytułu i etykiet
plt.title('Średnie wartości ciśnienia krwi według wieku i płci', fontsize=16)
plt.xlabel('Wiek', fontsize=14)
plt.ylabel('Średnia wartość (mmHg)', fontsize=14)

# Dodanie legendy
plt.legend(title='Płeć')

# Dodanie siatki dla lepszej czytelności
plt.grid(True)

# Wyświetlenie wykresu
plt.show()

# Wykres słupkowy dla cholesterolu
plt.figure(figsize=(18, 8))
sns.barplot(data=data, x='Age', y='Cholesterol (mg/dL)', hue='Gender', palette='tab10', errorbar=None)

# Dodanie tytułu i etykiet
plt.title('Średnie wartości cholesterolu według wieku i płci', fontsize=16)
plt.xlabel('Wiek', fontsize=14)
plt.ylabel('Średnia wartość (mg/dL)', fontsize=14)

# Dodanie legendy
plt.legend(title='Płeć')

# Dodanie siatki dla lepszej czytelności
plt.grid(True)

# Wyświetlenie wykresu
plt.show()
```
