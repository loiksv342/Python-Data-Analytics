```python
# import pandas as pd
# df = pd.read_csv('/kaggle/input/healthcare-dataset-2019-2024/healthcare_dataset.csv')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from math import pi
from matplotlib.sankey import Sankey
```

```python
df = pd.read_csv(r'healthcare_dataset.csv')
df.info()
```

```python
df.head(4)
```

### Formatting columns for better visibility


```python
df['Name'] = df['Name'].apply(lambda x: x.title())
df['Billing Amount'] = df['Billing Amount'].round(2)
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'], errors='coerce')
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], errors='coerce')


df.head(5)
```

### Question 1: What is the distribution of blood types by gender?

```python
count_df = df.groupby(['Gender', 'Blood Type']).size().reset_index(name='Count')


plt.figure(figsize=(10, 6))
sns.barplot(x='Blood Type', y='Count', hue='Gender', data=count_df, palette='Set2')

plt.title('Distribution of Blood Types by Gender')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

### Question 2: What is the distribution of blood types in the dataset

```python
blood_type_counts = df['Blood Type'].value_counts()

# Apply Seaborn style
sns.set(style="whitegrid")

# Define colors
custom_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Example colors

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    blood_type_counts,
    labels=blood_type_counts.index,
    autopct=lambda p: f'{p:.1f}%',  
    startangle=140,
    colors=custom_colors,
    wedgeprops=dict(edgecolor='#000')  # Optional: Add edge color to improve visibility
)

plt.title('Distribution of Blood Types')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle

plt.show()
```

### Question 3: What is the age distribution of patients

```python
sns.set(style="whitegrid")

number_of_bins = 15

plt.figure(figsize=(12, 8))
plt.hist(df['Age'].dropna(), bins=number_of_bins, color='#1f77b4', edgecolor='black')

plt.title('Distribution of Patient Ages', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

### Question 4: What are the average billing amounts by hospital? Top 5

```python
average_billing = df.groupby('Hospital')['Billing Amount'].mean().reset_index()

# Posortuj dane według średnich kosztów w malejącej kolejności
sorted_billing = average_billing.sort_values(by='Billing Amount', ascending=False)

# Wybierz top 5 szpitali z najwyższymi średnimi kosztami
top_5_hospitals = sorted_billing.head(5).copy()

# Konwersja kolumny 'Billing Amount' na int32
top_5_hospitals['Billing Amount'] = top_5_hospitals['Billing Amount'].astype('int32')

# Przygotowanie danych do wykresu
top_5_hospitals.set_index('Hospital', inplace=True)

# Tworzenie wykresu słupkowego z kolorami
plt.figure(figsize=(18, 5))
ax = top_5_hospitals.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

# Dodanie wartości nad słupkami
ax.bar_label(ax.containers[0], labels=top_5_hospitals['Billing Amount'].astype(str))

# Dodanie tytułu i etykiet osi
plt.title('Stacked Bar Plot of Average Billing Amounts for Top 5 Hospitals')
plt.xlabel('Hospital')
plt.ylabel('Average Billing Amount')
plt.grid(True)
plt.legend().set_visible(False)
plt.show()
```

### Question 5: What are the age distributions of patients by blood type?


```python
plt.figure(figsize=(16,8))

sns.histplot(data = df, x = 'Age', hue = 'Blood Type', multiple = 'stack')
plt.xlabel('Patient Count')
plt.ylabel('Age')
```

```python
condition_counts = df.groupby(['Blood Type', 'Medical Condition']).size().unstack(fill_value=0)

# Prepare data for visualization
condition_counts = condition_counts.reset_index()
condition_counts = condition_counts.melt(id_vars='Blood Type', var_name='Medical Condition', value_name='Count')

# Visualization
plt.figure(figsize=(12, 8))
sns.barplot(data=condition_counts, x='Medical Condition', y='Count', hue='Blood Type')

# Adjust legend
plt.legend(title='Blood Type', loc='upper left', bbox_to_anchor=(1, 1), ncol=1)

plt.title('Count of Medical Conditions by Blood Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Question 6: What are the distributions of medical conditions by blood type

```python
grouped_data = df.groupby(['Blood Type', 'Medical Condition']).size().reset_index(name='Count')

# Wykres Scatterplot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=grouped_data, x='Blood Type', y='Count', hue='Medical Condition', palette='viridis', s=100)
plt.title('Scatterplot of Medical Conditions by Blood Type')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.legend(title='Medical Condition')
plt.show()
```

### Question 7: What is the distribution of patients' length of stay?

```python
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

mean_length_of_stay = df['Length of Stay'].mean()
plt.figure(figsize=(14, 6))
sns.histplot(df['Length of Stay'].dropna(), bins=30, kde=True)
plt.title('Distribution of Length of Stay in Hospital')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()



```

### Question 8: How does the median length of stay compare to the overall spread?

```python
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
median_length_of_stay = df['Length of Stay'].median()
plt.figure(figsize=(14, 6))
sns.boxplot(x=df['Length of Stay'])
plt.title('Boxplot of Length of Stay in Hospital')
plt.xlabel('Length of Stay (days)')
plt.grid(True)
plt.show()
```
